#Label encoding

```python
categ =  ['Country', 'Status', 'Life expectancy ', 'Adult Mortality',
       'Alcohol','Total expenditure','GDP',
       'Population', ' thinness  1-19 years',
       'Income composition of resources', 'Schooling']

from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for i in categ:
  le.fit(df[i])
  le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
  print(le_name_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

categ = ['Country', 'Year', 'Status', 'Life expectancy ', 'Adult Mortality',
       'Alcohol','Total expenditure','GDP',
       'Population', ' thinness  1-19 years',
       'Income composition of resources', 'Schooling']

df[categ] = df[categ].apply(le.fit_transform)
```

```python
X = df[['Country', 'Year', 'Status','Adult Mortality',
       'Alcohol','Total expenditure','GDP',
       'Population', ' thinness  1-19 years',
       'Income composition of resources', 'Schooling']]
Y =  df['Life expectancy ']
```

```python
from sklearn.preprocessing import StandardScaler
```

#Standardizing the data

```python
#Standardize the dataset using standard scaler
scaler = StandardScaler()
scaler.fit(X)
scaled = scaler.transform(X)
```

#Train Test And Split of the data

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

#Implementing Linear Regression Model

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print('Intercept =', model.intercept_)
```

```python
print('Coefficients = ', model.coef_)
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
r2score = r2_score(y_test, y_pred)
r2score
```

#Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.35,random_state=0)
```

```python
forest_reg = RandomForestRegressor()
forest_reg.fit(x_train, y_train)
```

```python
forest_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
y_pred=forest_reg.predict(x_test)
forest_reg.score(x_test,y_test)
```
